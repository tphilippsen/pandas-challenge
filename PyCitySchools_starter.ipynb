```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

school_data_complete
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# calcualte total schools
total_schools = len(school_data_complete["school_name"].unique())
total_schools 
```

```python
# calculate total students
total_students = school_data_complete['student_name'].count()
total_students
```

```python
# calculate total budget
unique_budget = school_data_complete['budget'].unique()

total_budget = unique_budget.sum()
total_budget
```

```python
# calculate avg math score

avg_math_score = school_data_complete['math_score'].mean()
avg_math_score
```

```python
# calculate avg reading score
avg_reading_score = school_data_complete['reading_score'].mean()
avg_reading_score
```

```python
# calculate % of students with math score of 70 or higher
math_pass = len(school_data_complete.loc[school_data_complete['math_score'] >= 70])
pct_math_pass = (math_pass/total_students) * 100
pct_math_pass
```

```python
# calculate % of students with reading score 70 or higher
read_pass = len(school_data_complete.loc[school_data_complete['reading_score'] >= 70])
pct_read_pass = (read_pass/total_students) * 100
pct_read_pass
```

```python
# calculate % of students with reading and math scores 70 or greater
read_math_pass = len(school_data_complete.loc[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70), :])
percent_pass_scores = (read_math_pass/total_students) *100
percent_pass_scores

```

```python
summary_df = pd.DataFrame({'Total Schools' : [total_schools],
                          "Total Students": [total_students],
                          "Total Budget": [total_budget],
                          "Average Math Score": [avg_math_score],
                          "Average Reading Score": [avg_reading_score],
                          "% Passing Math": [pct_math_pass],
                          "% Passing Reading": [pct_read_pass],
                          "% Overall Passing": [percent_pass_scores]})
summary_df
```

## School Summary

```python
# school by type
school_type = school_data.set_index(['school_name'])['type']
school_type
```

```python
# total students by school
students_school = school_data_complete['school_name'].value_counts()
students_school
```

```python
# total budget by school
school_budget = school_data.groupby(['school_name']).mean()['budget']
school_budget
```

```python
# per student budget
per_student_budget = (school_budget) / (students_school)
per_student_budget
```

```python
# avg reading score by school
avg_read_score = student_data.groupby('school_name').mean()['reading_score']

avg_read_score

```

```python
# avg math score by school
avg_math_score = student_data.groupby('school_name').mean()['math_score']

avg_math_score
```

```python
# % passing reading
total_pass_read = school_data_complete.loc[(school_data_complete['reading_score'] >= 70)]
per_pass_read = total_pass_read.groupby(['school_name'])['student_name'].count()
student_pass_read = ((per_pass_read) / (students_school)) * 100
student_pass_read
```

```python
# % passing math
total_pass_math = school_data_complete.loc[(school_data_complete['math_score']>= 70)]
per_pass_math = total_pass_math.groupby(['school_name'])['student_name'].count()
student_pass_math = ((per_pass_math) / (students_school)) * 100
student_pass_math
```

```python
# % passing math and reading
total_pass_read_math = school_data_complete.loc[(school_data_complete ['reading_score']>=70) & (school_data_complete['math_score']>=70)]
per_pass_read_math = total_pass_read_math.groupby(['school_name'])['student_name'].count()
student_pass_read_math = ((per_pass_read_math) / (students_school)) * 100
student_pass_read_math
```

```python
# put info in a dataframe
school_data_summary = pd.DataFrame({'School Type' : school_type,
                                    'Total Students' : students_school,
                                    'Total School Budget' : school_budget,
                                    'Per Student Budget' : per_student_budget,
                                    'Average Math Score': avg_math_score,
                                    'Average Reading Score': avg_read_score,
                                    '% Passing Math': student_pass_math,
                                    '% Passing Reading' : student_pass_read,
                                    '% Overall Passing' : student_pass_read_math})
school_data_summary
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Binning
bin = [0 ,585, 630, 645, 680]
group_name = ['<$585', '$585-630', '$630-645' , '$645-680' ]

school_data_summary['Spending Ranges (Per Student)'] = pd.cut(school_data_summary['Per Student Budget'], bin, labels=group_name, include_lowest=True)
student_spend = school_data_summary.groupby('Spending Ranges (Per Student)').mean()
student_spend.drop(['Total Students', 'Total School Budget', 'Per Student Budget'], axis=1)
```

## Scores by School Size
